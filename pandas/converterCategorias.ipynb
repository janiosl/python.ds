Carga dos dados usando Panda

```python
import pandas as pd
dados = pd.read_excel(arquivo)
```

```python
dados.head()
```

Criação de uma nova coluna com a variável categória

```python
import sys
```

```python
dados['alturaCat']=pd.cut(
        dados['Altura'],
        bins=[0, 1.5, 1.8, sys.maxsize],
        labels=['Baixo', 'Médio', 'Alto']
     )
```

```python
dados
```

Criação de outra coluna com as categorias em formato numérico (inteiro) para uso em algoritmos

```python
dados['alturaCat2']=pd.cut(
    dados['Altura'],
    bins=[0, 1.5, 1.8, sys.maxsize],
    labels=[0,1,2]
)
```
